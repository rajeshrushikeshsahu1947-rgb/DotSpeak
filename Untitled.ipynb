```python
import os
import shutil
import random

# Paths
dataset_dir = r"C:/Users/Rajesh/braille10/Braille Alphabet Image Dataset (A–Z)"
output_dir = r"C:/Users/Rajesh/braille10/split_dataset"

train_dir = os.path.join(output_dir, "train")
val_dir   = os.path.join(output_dir, "val")

# Create train/val directories
for split in [train_dir, val_dir]:
    if not os.path.exists(split):
        os.makedirs(split)

# Loop through each class folder (A–Z)
for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)
    if not os.path.isdir(class_path):
        continue

    # Make class subfolders in train/val
    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)

    # Get all images in this class
    images = os.listdir(class_path)
    random.shuffle(images)

    # 80% train, 20% val
    split_idx = int(0.8 * len(images))
    train_images = images[:split_idx]
    val_images   = images[split_idx:]

    # Copy files
    for img in train_images:
        shutil.copy(os.path.join(class_path, img),
                    os.path.join(train_dir, class_name, img))
    for img in val_images:
        shutil.copy(os.path.join(class_path, img),
                    os.path.join(val_dir, class_name, img))

print("✅ Dataset split complete!")
print(f"Train folder: {train_dir}")
print(f"Val folder: {val_dir}")
```

```python
def extract_braille_pattern(img_path):
    img = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE)
    _, binary = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
    inverted = cv2.bitwise_not(binary)

    # Find contours (dots)
    contours, _ = cv2.findContours(inverted, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    centers = []
    for cnt in contours:
        x,y,w,h = cv2.boundingRect(cnt)
        if 5 < w < 20 and 5 < h < 20:  # filter small blobs
            cx, cy = x + w//2, y + h//2
            centers.append((cx, cy))

    if not centers:
        return [0,0,0,0,0,0]

    # Normalize positions into 2x3 grid
    # Sort by x (left/right) then y (top/mid/bottom)
    centers = sorted(centers, key=lambda p: (p[0], p[1]))

    # Split into left and right columns
    xs = [c[0] for c in centers]
    ys = [c[1] for c in centers]

    # Compute thresholds
    x_mid = (min(xs) + max(xs)) / 2
    y_thirds = np.linspace(min(ys), max(ys), 4)  # 3 rows

    pattern = [0,0,0,0,0,0]  # [Ltop,Lmid,Lbot,Rtop,Rmid,Rbot]

    for cx, cy in centers:
        col = 0 if cx < x_mid else 1
        if cy < y_thirds[1]:
            row = 0
        elif cy < y_thirds[2]:
            row = 1
        else:
            row = 2
        idx = row + col*3
        pattern[idx] = 1

    return pattern
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import matplotlib.pyplot as plt
from IPython.display import display, clear_output

# Load your trained YOLO classification model (A–Z letters)
cls_model = YOLO(r"C:\Users\Rajesh\braille10\runs\classify\train-5\weights\best.pt")

# --- Preprocessing ---
def preprocess(frame):
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    # Adaptive thresholding (better for real camera input)
    binary = cv2.adaptiveThreshold(gray, 255,
                                   cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                   cv2.THRESH_BINARY, 11, 2)
    # Invert so dots are black
    inverted = cv2.bitwise_not(binary)
    # Morphological cleaning
    kernel = np.ones((3,3), np.uint8)
    clean = cv2.morphologyEx(inverted, cv2.MORPH_OPEN, kernel)
    return clean

# --- Extract ROI (manual or largest contour) ---
def extract_roi(clean):
    contours, _ = cv2.findContours(clean, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if len(contours) == 0:
        return None
    # Take largest contour as ROI
    c = max(contours, key=cv2.contourArea)
    x,y,w,h = cv2.boundingRect(c)
    roi = clean[y:y+h, x:x+w]
    return roi

# --- Classify single ROI ---
def classify_roi(roi):
    if roi is None or roi.size == 0:
        return ""
    # Resize to match training input size
    roi_resized = cv2.resize(roi, (64,64))
    cv2.imwrite("temp_cell.jpeg", roi_resized)
    results = cls_model.predict(source="temp_cell.jpeg", verbose=False)
    probs = results[0].probs.data.cpu().numpy()
    letter = chr(np.argmax(probs) + 65)  # A–Z
    return letter

# --- Jupyter-friendly frame display ---
def show_frame(frame, text=""):
    if text:
        cv2.putText(frame, text, (50, 50), cv2.FONT_HERSHEY_SIMPLEX,
                    1, (0, 255, 0), 2)
    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb)
    plt.axis("off")
    clear_output(wait=True)
    display(plt.gcf())
    plt.close()

# --- Real-time loop ---
def realtime_braille_letter(max_frames=200):
    cap = cv2.VideoCapture(0)  # webcam
    frame_count = 0
    while frame_count < max_frames:
        ret, frame = cap.read()
        if not ret:
            break

        clean = preprocess(frame)
        roi = extract_roi(clean)
        letter = classify_roi(roi)

        # Show result inline
        show_frame(frame, letter)

        frame_count += 1

    cap.release()

# --- Run ---
realtime_braille_letter()
```
