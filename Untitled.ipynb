```python
import os
import shutil
import random

# Paths
dataset_dir = r"C:/Users/Rajesh/braille10/Braille Alphabet Image Dataset (A–Z)"
output_dir = r"C:/Users/Rajesh/braille10/split_dataset"

train_dir = os.path.join(output_dir, "train")
val_dir   = os.path.join(output_dir, "val")

# Create train/val directories
for split in [train_dir, val_dir]:
    if not os.path.exists(split):
        os.makedirs(split)

# Loop through each class folder (A–Z)
for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)
    if not os.path.isdir(class_path):
        continue

    # Make class subfolders in train/val
    os.makedirs(os.path.join(train_dir, class_name), exist_ok=True)
    os.makedirs(os.path.join(val_dir, class_name), exist_ok=True)

    # Get all images in this class
    images = os.listdir(class_path)
    random.shuffle(images)

    # 80% train, 20% val
    split_idx = int(0.8 * len(images))
    train_images = images[:split_idx]
    val_images   = images[split_idx:]

    # Copy files
    for img in train_images:
        shutil.copy(os.path.join(class_path, img),
                    os.path.join(train_dir, class_name, img))
    for img in val_images:
        shutil.copy(os.path.join(class_path, img),
                    os.path.join(val_dir, class_name, img))

print("✅ Dataset split complete!")
print(f"Train folder: {train_dir}")
print(f"Val folder: {val_dir}")
```

```python
import os

# Paths
train_dir = r"C:/Users/Rajesh/braille10/split_dataset/train"
val_dir   = r"C:/Users/Rajesh/braille10/split_dataset/val"

output_images_train = r"C:/Users/Rajesh/braille10/braille_dataset/images/train"
output_images_val   = r"C:/Users/Rajesh/braille10/braille_dataset/images/val"
output_labels_train = r"C:/Users/Rajesh/braille10/braille_dataset/labels/train"
output_labels_val   = r"C:/Users/Rajesh/braille10/braille_dataset/labels/val"

# Create folders
for d in [output_images_train, output_images_val, output_labels_train, output_labels_val]:
    os.makedirs(d, exist_ok=True)

# Map class names to IDs
class_names = sorted(os.listdir(train_dir))
class_to_id = {cls: idx for idx, cls in enumerate(class_names)}

def process_split(split_dir, out_img_dir, out_lbl_dir):
    for cls in os.listdir(split_dir):
        cls_path = os.path.join(split_dir, cls)
        if not os.path.isdir(cls_path):
            continue
        for img in os.listdir(cls_path):
            if img.lower().endswith((".jpg",".png",".jpeg")):
                src = os.path.join(cls_path, img)
                dst = os.path.join(out_img_dir, img)
                # Copy image
                os.system(f'copy "{src}" "{dst}"')
                # Write label file
                label_file = os.path.join(out_lbl_dir, img.rsplit(".",1)[0]+".txt")
                with open(label_file, "w") as f:
                    f.write(f"{class_to_id[cls]} 0.5 0.5 1.0 1.0\n")

process_split(train_dir, output_images_train, output_labels_train)
process_split(val_dir, output_images_val, output_labels_val)

print("✅ YOLO images + labels created")
```
